```python
import tensor as tn
```

```python
A = tn.Tensor((3,4))
```

```python
for i in range(12):
    x = (i+1)**2;
    A._ptr.contents.data[i] = x;
    
```

```python
A.print()
```

```python
B = tn.Tensor((3,4));
```

```python
for i in range(12):
    x = -1 *(i+1)**2;
    B._ptr.contents.data[i] = x;
```

```python
B.print()
```

```python
C = A + B
```

```python
C.print()
```
